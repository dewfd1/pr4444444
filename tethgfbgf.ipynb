Цель: Необходимо использовать датасет с прошлой практической работы и визуализировать данные в виде графиков: Линейный график - минимум 3 штуки (можно использовать наложение графиков), Столбчатая диаграмма- минимум 3 штуки (можно также использовать наложение), Круговая диаграмма - минимум 2 штуки, Диаграмма рассеяния - минимум 2 штуки, Гистограмма - минимум 3 штуки

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Импортируем нужные библиотеки. Рandas предоставляет инструменты для обработки и анализа структурированных данных. Библиотека pandas также предоставляет функциональность для чтения и записи данных из различных источников, включая CSV. Библиотека matplotlib.pyplot, предоставляет функции для создания различных видов графиков и диаграмм: линейные графики, столбчатые диаграммы, круговые диаграммы, диаграммы рассеяния, гистограммы.

```python
data = pd.read_csv('roblox_account_index.csv')
```

Загрузка данных из CSV-файла.

```python
plt.figure(figsize=(10, 6))
plt.plot(data['user_id'], label='user_id')
plt.plot(data.index, data['username'], label='username')
plt.plot(data.index, data['avatar'], label='avatar')
plt.title('Линейные графики')
plt.xlabel('Индекс')
plt.ylabel('Значение')
plt.legend()
plt.show()
```

Линейные графики отображают изменение значений пользовательских идентификаторов (user_id), имени пользователя (username) и аватаров (avatar) по индексу. Можно увидеть, как меняются эти значения в зависимости от порядкового номера записи.

```python
plt.figure(figsize=(10, 6))
plt.bar(data.index[:3], data['user_id'][:3], label='user_id')
plt.bar(data.index[:3], data['premium'][:3], bottom=data['user_id'][:3], label='premium')
plt.bar(data.index[:3], data['avatar'][:3], bottom=data['user_id'][:3]+data['premium'][:3], label='avatar')
plt.title('Столбчатые диаграммы')
plt.xlabel('Индекс')
plt.ylabel('Значение')
plt.legend()
plt.show()
```

Столбчатые диаграммы позволяют сравнить значения user_id, premium и avatar для первых трех записей в файле. Они помогают увидеть различия в значениях между этими параметрами.

```python
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.pie(data['username'], labels=data['username'])
plt.title('Круговая диаграмма 1')
plt.subplot(122)
plt.pie(data['premium'], labels=data['premium'])
plt.title('Круговая диаграмма 2')
```

Круговая диаграмма для premium отображает соотношение премиум статуса между двумя пользователями, показывая, какая доля у каждого из них.

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['user_id'][:2], data['premium'][:2], label='user_id vs premium')
plt.scatter(data['user_id'][2:4], data['premium'][2:4], label='user_id vs premium')
plt.title('Диаграммы рассеяния')
plt.xlabel('user_id')
plt.ylabel('premium')
plt.legend()
plt.show()
```

Диаграммы рассеяния позволяют увидеть взаимосвязь между user_id и premium для двух пар пользователей. Они помогают определить, существует ли какая-либо зависимость между этими параметрами.

```python
plt.figure(figsize=(10, 5))
plt.subplot(231)
plt.hist(data['user_id'], bins=10)
plt.title('Гистограмма 1')
plt.subplot(232)
plt.hist(data['username'], bins=10)
plt.title('Гистограмма 2')
plt.subplot(233)
plt.hist(data['avatar'], bins=10)
plt.title('Гистограмма 3')
```

Гистограммы показывают распределение значений user_id, username и avatar , помогая увидеть, какие значения наиболее часто встречаются.
